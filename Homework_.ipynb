```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'City': ['London', 'Paris', 'Tokyo'],
    'Population': [9.0, 2.1, 37.4]
}

df = pd.DataFrame(data)
plt.bar(df["City"], df["Population"]) #Plot city name and population on X axis
plt.xlabel("City") # Label for X axis
plt.ylabel("Population") #Label for Y axis
plt.title("Population of Cities") # Title for cities
plt.show() #Display output
```

# Task 5

```python
import matplotlib.pyplot as plt  # Import the plotting library

# Defining x and y values
x = [0, 1, 2, 3]
y = [0, 1, 4, 9]

# Creating the line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')

# Adding title and axis labels
plt.title("Square Numbers Plot")          # Title of the plot
plt.xlabel("X values")                    # Label for x-axis
plt.ylabel("Y values (x squared)")        # Label for y-axis

# Adding a grid for better readability
plt.grid(True)

# Saving the figure as an image file
plt.savefig("squareplot.png")

plt.show()
```
