##OPENING AND READING A FILE


```python
f = open("test.txt", "r")
print(f.read())
f.close()
```

ReadLine Function


```python
f = open("test.txt", "r")
print(f.readlines())
f.close()
```

Read Function

```python
f = open("test.txt", "r")
print(f.read(6))
f.close()
```

```python
f = open("test.txt", "r")
print(f.readline(128))
f.close()
```

WITH OPERATION

```python
with open("test.txt", "r") as f:
  print(f.read())
```

```python
f = open("test1.txt", "w")
f.write("I am writing to this file")
f.close()
```

Using With


```python
with open("test.txt", "w") as f:
    print(f.write("Hello, world!"))
```

APPENDING TO A FILE

```python
f = open("test.txt", "a")
f.write("\n I am appending to this file")
f.close()
```
