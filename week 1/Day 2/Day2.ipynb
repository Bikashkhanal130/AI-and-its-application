Dictionaries in Python


```python
# Creating a Dictionary

student = {
    "name" : "Hari Lal",
    "age" : 27,
    "course" : "Computer Science"

}

#Accessing values of a particular key in a dictionary

print(student["age"])

#Updating values of a particular key in as dictionary
student["age"] = 22

print(student["age"])


#Adding new key-value pair on the Dictionary
student["address"] = "Kathmandu"
print(student["address"])

print (student)


#removing a key-value pair from Dictionary

del student["course"]
print(student)

```

Simple Interest Calculation

```python
# Taking Input from users
principal = float(input("Enter Principal Amount"))
rate = float(input("Enter Rate"))
time = float(input("Enter Time"))
#Using formula to calculate simple interest
simple_interest = (principal * rate * time) / 100
#printing output
print("The simple interest is " , simple_interest)
```

```python
# Calculation of BMI
```

```python

#taking input from user
weight = float(input("Enter your weight"))
height = float(input("Enter your height"))

#calculating BMI using its formula
BMI = weight / (height * height)

#printing output
print("Your BMI is", BMI)

#Determining weight category

if BMI <= 18.5:
  print("You are underweight")
elif BMI <= 24.9:
    print("You are normal_weight")
elif BMI <= 29.9:
      print("You are overweight")
else:
      print("You are obese")
```
