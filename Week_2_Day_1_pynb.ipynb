# Exercise 5: Basic Plotting with Matplotlib

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

x = [1, 1, 2, 3, 3]
y = [5, 2, 1, 2, 5]

x = [-4, -3, -2, -1, 0, 1, 2, 3, 4]
y = [16, 9, 4, 1, 0, 1, 4, 9, 16]

plt.plot(x, y, marker='o')
plt.title("Basic Plotting with Matplotlib")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Half U Shape

```python
import matplotlib.pyplot as plt

# Half U shape
x = [-4, -3, -2, -1, 0]
y = [16, 9, 4, 1, 0]

plt.plot(x, y, marker='o')
plt.title("Half U Shape")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

# Exercise 6: Visualizing a DataFrame with Matplotlib

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [25, 30, 35, 40, 45],
}

df = pd.DataFrame(data)

plt.bar(df["Name"],df ["Score"])
plt.title("Visualizing a DataFrame with Matplotlib")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```
